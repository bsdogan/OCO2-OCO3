## Obj: Creating animation with frames

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib import animation
from mpl_toolkits.basemap import Basemap
from mpl_toolkits import basemap

# LIST the path to files
import xarray
import netCDF4 as nc4
import glob
import pandas as pd
from datetime import datetime


from IPython.display import display, HTML, Image
```

### Reading the OCO2 data for 16 days

```python
# DATE time function
def conv_date(d):
    return datetime.strptime(str(d), '%Y%m%d%H%M%S%f')
```

```python
# netcdf file list
netcdf_files= []

total_files= []

# gather the data from each columns
xco2= []
long= []
lat= []
dateTime= []
xco2_qual= []

# READING multiple files (days) of oco2    
for f in glob.glob('oco2_data/*nc4'): #, recursive= True)

    # Create dataframe
    netcdf_files.append(f)
    data= nc4.Dataset(f, "r")
    
    xco2.append(data["xco2"][:])
    lat.append(data["latitude"][:])
    long.append(data["longitude"][:])
    dateTime.append(data["sounding_id"][:])
    xco2_qual.append(data["xco2_quality_flag"][:])
    
```

```python
df_xco2_new= pd.DataFrame(columns=[ "Latitude", "Longitude", "DateTime","xco2", "xco2_quality_flag"])

# for j in range(0, len(xco2)):
    
df_xco2_new["xco2"]= np.concatenate([xco2[j] for j in range(0, len(xco2))])
df_xco2_new["Latitude"]= np.concatenate([lat[j] for j in range(0, len(lat))])
df_xco2_new["Longitude"]= np.concatenate([long[j] for j in range(0, len(long))])
df_xco2_new["DateTime"]= np.concatenate([dateTime[j] for j in range(0, len(dateTime))])
df_xco2_new["xco2_quality_flag"]= np.concatenate([xco2_qual[j] for j in range(0, len(xco2_qual))])


# DATE TIME columns:
#df_xco2_new['DateTime']= pd.to_datetime(df_xco2_new['DateTime'])


#df_xco2_new['DateTime']= df_xco2_new['DateTime'].astype(str)

df_xco2_new['DateTime']= df_xco2_new['DateTime'].apply(conv_date)

df_xco2_new['Year']= df_xco2_new['DateTime'].dt.year
df_xco2_new['Month']= df_xco2_new['DateTime'].dt.month
df_xco2_new['Day']= df_xco2_new['DateTime'].dt.day



df_xco2_new
```

```python
filtered_xco2= df_xco2_new[:40000]
```
